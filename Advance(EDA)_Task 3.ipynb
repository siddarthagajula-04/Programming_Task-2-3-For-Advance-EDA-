## Maximum Product of Two Elements in an Array

```python
def maxProduct(nums):
    max_val = 0
    n = len(nums)

    for i in range(n):
        for j in range(i + 1, n):
            val = (nums[i]-1)*(nums[j]-1)
            if val > max_val:
                max_val = val
    return max_val
```

```python
nums = [3,4,5,2]
maxProduct(nums)
```

```python
nums = [1,5,4,5]
maxProduct(nums)
```

```python
nums = [3,7]
maxProduct(nums)
```

## Count Number of Teams

```python
def numTeams(rating):
    count = 0
    n = len(rating)

    for i in range(n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):
                if (rating[i] < rating[j] < rating[k]) or \
                   (rating[i] > rating[j] > rating[k]):
                    count += 1

    return count
```

```python
rating = [2,5,3,4,1]
numTeams(rating)
```

```python
rating = [2,1,3]
numTeams(rating)
```

```python
rating = [1,2,3,4]
numTeams(rating)
```

## Number of Students Doing Homework at a Given Time

```python
def busyStudent(startTime, endTime, queryTime):

    count = 0
    for i in range(len(startTime)):
        if startTime[i] <= queryTime <= endTime[i]:
            count += 1
    
    return count
```

```python
busyStudent([1,2,3],[3,2,7],4)
```

```python
busyStudent([4],[4],4)
```

## Number of Steps to Reduce a Number to Zero

```python
def numberOfSteps(num):
    Steps = 0
    
    while num > 0:
        if num % 2 == 0:
            num //= 2

        else:
            num -= 1
        Steps += 1

    return Steps
```

```python
num = 14
numberOfSteps(num)
```

```python
num = 8
numberOfSteps(num)
```

```python
num = 123
numberOfSteps(num)
```

## Counting Bits

```python
def countBits(n):
    ans = [0] * (n + 1)

    for i in range(1, 1 + n):
        ans[i] = ans[i // 2] + (i % 2)

    return ans
```

```python
n = 2
countBits(n)
```

```python
n = 5
countBits(n)
```
